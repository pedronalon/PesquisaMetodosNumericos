```python
from sympy import symbols, diff, sympify, sin,cos,exp, integrate, lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = e^x sen(x) 
# g(x) = e^x sen(x) + e^x cos(x) - Derivada exata 
x = symbols('x')
f =  sin(x)*exp(x)
f_sympy = sympify(f)
derivada = diff(f_sympy,x)
integral = integrate(f_sympy,x)
print(derivada)
print(integral)

f_numerica = lambdify(x, f, 'numpy')
derivada_numerica = lambdify(x, derivada, 'numpy')
integral_numerica = lambdify(x, integral, 'numpy')

xx = np.linspace(-2,2,100) 
y = f_numerica(xx)
dfdx = derivada_numerica(xx)
sum = integral_numerica(xx)


plt.figure(figsize=(10, 6))

# Plotagem
plt.plot(xx, y, label='Função original: $e^x \sin(x)$', color='blue')
plt.plot(xx, dfdx, label='Derivada: $e^x (\sin(x) + \cos(x))$', color='red', linestyle='--')
plt.plot(xx, sum, label='Integral: $-\\frac{e^x \cos(x)}{2} + \\frac{e^x \sin(x)}{2}$', color='green', linestyle=':')

# Configurações do gráfico
plt.title('Função, Derivada e Integral')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

```
