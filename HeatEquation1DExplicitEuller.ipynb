```python
import numpy as np 
from numpy import sin, pi
import matplotlib.pyplot as plt
```

```python
c = 0.1 # coeficiente de difusão
n = 20 # número de divisões da barra
L = 1.0 # comprimento da barra
h = L/n # tamanho do passo no espaço 
dt = 0.01 # tamanho do passo no tempo
t_final = 1
#condições de contorno (cada extremidade da barra)
T_left = 0 
T_right = 0

def g(x):
    return sin(pi*x)

tempo = np.arange(0, t_final, dt) #espaçamento fisico de t = 0.01, arange para um vetor em um intervalo de de 0 a t_final com passo t
x = np.linspace(0, L, n) # quantidade n de pontos no espaço, linspace para um vetor num intervalo com n pontos

T = np.zeros(n) 
T_new = np.zeros(n)

T[0] = T_left
T[-1] = T_right

T_new[0] = T_left
T_new[-1] = T_right

#T[int(n/2)] = 100 # pertubação no sistema de 100 graus no meio da barra 
T[:] = g(x[:])



plt.figure(figsize=(12,8))
plt.plot(x,T)
plt.grid(True)
plt.show() 




```

```python

plt.figure(figsize=(12,8))

for j in range(len(tempo)): #looping do tempo, no tamanho do vetor de tempo
    for i in range(1, n-1): #looping do espaço, exceto nas extremidades
        T_new[i] = T[i] + dt*(c*(T[i+1]-2*T[i]+T[i-1])/(h**2))  
    
    if(j%10 == 0):
        plt.plot(x,T,label = f'tempo = {tempo[j]:.2f}')
     
    
    T = T_new # atualização do vetor
    
plt.legend()   
plt.grid(True)
plt.show()

```
