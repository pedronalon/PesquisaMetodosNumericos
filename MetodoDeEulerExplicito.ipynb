```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def f(y):
    return 10*y

def f_exato(x):
    return np.exp(10*x)

def g(y):
    return (-10)*y

def g_exato(x):
    return np.exp((-10)*x)
```

```python
n = 1000
t = np.linspace(0,1,n)
y = np.zeros(n)
y[0] = 1
h = t[1] - t[0]
for i in range(n-1):
    y[i+1] = y[i]+ (h*f(y[i]))


plt.figure(figsize=(12,8))    
plt.plot(t,y,'-o', color='red')
plt.plot(t,f_exato(t),color='black')
plt.grid(True)
plt.show() 



```
