```python
import numpy as np
import matplotlib.pyplot as plt 

```

```python

def f_exato(x):
    return np.exp(10*x)

def g_exato(x):
    return np.exp((-10)*x)

def h(y):
    return 10*y*(1-y)

def h_extato(t):
    x = 0.1
    return  1/ (1 + ((1 - x) / x) * np.exp(-10 * t))


```

```python
n = 500
t = np.linspace(0,1,n)
y = np.zeros(n)
y[0] = 1 
h = t[1] - t[0] 
for i in range(n-1): 
    y[i+1] = y[i]/(1- 10*h)

plt.figure(figsize=(12,8))
plt.plot(t,f_exato(t), color="black")
plt.plot(t,y,'o',color="red")
plt.grid(True)
plt.show()
```

```python
n = 500 
t = np.linspace(0,1,n)
y = np.zeros(n)
y[0] = 1
h = t[1] - t[0]

for i in range(n-1):
    y[i+1] = y[i] / (1 - h *(-10))

plt.figure(figsize=(12,8))
plt.plot(t,g_exato(t), color = "black")
plt.plot(t,y,'o', color ="red")
plt.grid(True)
plt.show()
```

```python
n = 1000
t = np.linspace(0,5,n)
h = t[1]-t[0]
y = np.zeros(n)
y[0] = 0.1
a = 10

for i in range(n-1):
    A = h*a
    B = 1-h*a
    C = -y[i]
    delta = B**2 - 4*A*C

    y1 = (-B + np.sqrt(delta)) / (2*A)
    y2 = (-B - np.sqrt(delta))/ (2*A)

    if 0 <= y1 <= 1:
        y[i+1] = y1
    elif 0 <= y2 <= 1:
        y[i+1] = y2


plt.figure(figsize=(12,8))
plt.plot(t,h_extato(t), color="black")
plt.plot(t,y,'o',color="red")
plt.grid(True)
plt.show()
```
