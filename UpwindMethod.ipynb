```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# u_t + a u_x = 0, com a>0
a = 1.0
dt = 1.0
dx = 1.0
lmbd = a*dt/dx
m = 72
t = np.arange(0.0, 8.0, dt)
x = np.arange(-10.0, 11.0, dx)
u = np.zeros((len(x), len(t)))
#condição de contorno (?)
u[0,:] = 0 
u[-1,:] = 0
```

```python
def condicao_inicial(x):
    return np.exp(-x**2)


def solucao_exata(x,t,a):
    return np.exp(-(x - a * t)**2)


u[:,0] = condicao_inicial(x)

for i in range(0, len(t)-1):
    for j in range(1, len(x)):
        u[j,i+1] = (1-lmbd)*u[j,i] +lmbd*u[j-1,i]


cond = condicao_inicial(x)

plt.figure(figsize=(10,6))
plt.plot(x, u[:,0], label='t=0')
plt.plot(x, u[:,5], label='t=10')
plt.plot(x, u[:,7], label='t=20')
plt.plot(x, solucao_exata(x,t[7],a), label='t=20 (exata)', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
