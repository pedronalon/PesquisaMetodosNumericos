```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.01
b = a**2 - 2/3 * a**3
n = 3000
t = np.linspace(0,3,n)
h = t[1] - t[0]
y = np.zeros(n)
y[0] = 1
maxit = 10
def y_exato(t):
    return (0.01+t)/(np.sqrt(b+(2/3)*np.power(0.01+t,3)))

def f(y,t):
    return y/(a+t) - y**3

def F(y_novo,y_atual,t_novo,h):
    return y_novo - y_atual - h * (y_novo/(a+t_novo) - y_novo**3) 

def dF(y_novo,t_novo,h):
    return 1- h *(1/(a+t_novo) - 3*(y_novo**2))
```

```python

for i in range(n-1):
    t_novo = t[i+1]
    y_atual = y[i]

    chute = y_atual
    for j in range(maxit):
        y_calculado = F(chute,y_atual,t_novo,h)
        dy = dF(chute,t_novo,h)

        novoChute = chute - y_calculado/dy
        print(j,y_calculado)

        if abs(y_calculado) < 1e-8:
            chute = novoChute
            break

        chute = novoChute

    y[i+1] = chute




plt.figure(figsize=(12,8))
plt.plot(t,y_exato(t),color = "black")
plt.plot(t,y, 'o', color = "red")
plt.grid(True)
plt.show()
```
