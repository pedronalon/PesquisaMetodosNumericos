```python
import matplotlib.pyplot as plt 
import numpy as np 
from math import sin,cos,exp
```

```python
# f(x) = xÂ² + 4x + 4
# f'(x) = 2x + 4 Derivada Exata 

def f (x) :
    return 2*x + 4

def g (x) :
    return x**2 + 4*x + 4

h = 1
primeiro_ponto = True
for i in range (-10,10): 
    dfp = (g(i+h) - g(i))/h
    dfc = (g(i+h) - g(i-h))/2*h
    erro1 = abs(dfp - f(i))
    erro2 = abs(dfc - f(i))
    if primeiro_ponto:
        plt.scatter(i , dfp, color='red', label="Derivada progressiva", s=50, edgecolor='black', marker = "^")
        plt.scatter(i , dfc , color='blue', label="Derivadas central", s=50, edgecolor='black', marker = "D")
        primeiro_ponto = False
    else:
        plt.scatter(i , dfp, color='red', s=50, edgecolor='black', marker = "^")
        plt.scatter(i , dfc, color='blue', s=50, edgecolor='black', marker = "D")
    print("Erro da derivada progressiva:",erro1,"ponto:",i)
    print("Erro da derivada central:",erro2,"ponto:",i)

x = np.linspace (-10,10,200) 
plt.plot(x,f(x),color = "black", label = "Derivada exata de f")
plt.title("Derivada de f(x) por aproximaÃ§Ã£o com h = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend() 
plt.show()
 
```
