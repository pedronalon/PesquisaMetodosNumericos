```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# u_t + a u_x = 0, com a>0
a = 0.8
xi = -10.0
xf = 10.0

N = 202
M = 101

ti = 0.0 
tf = 7.0
#h = 0.04

dx = (xf - xi) /N


x = np.linspace(xi, xf, N)
t = np.linspace(ti, tf, M)

h = t[1] - t[0]

lmbd = a * h / dx

u = np.zeros((len(t), len(x)))
lmbd
```

```python
def condicao_inicial(x):
    return np.where((x >= 0) & (x <= 1), 1.0, 0.0)


def solucao_exata(x,t,a):
    return condicao_inicial(x-a*t)



u[0,:] = condicao_inicial(x)

# u = np.zeros(len(x))
# u = condicao_inicial(x)

# for i in range(len(t)-1):
#    u_new = np.zeros(len(x))
#    u_new[1:] = (1-lmbd)*u[1:] + lmbd*u[:-1]
#    u_new[0] = u_new[1]
#    u = u_new



for i in range(0,len(t)-1):
    u[i+1,1:] = (1-lmbd)*u[i,1:] + lmbd*u[i,:-1]
    
    # condição de contorno (?)
    u[i+1, 0] = 0.0
      
```

```python
for i in range(len(t)):
    if i%10 == 0:
        plt.figure(figsize=(10,4))
        plt.plot(x, u[i,:])
        plt.plot(x, solucao_exata(x,t[i],a), label='exata', linestyle='--')
        plt.title("solução em t = {}".format(t[i]))
        plt.legend()
        plt.grid(True)
        plt.show()
```
