```python
import jax 
import jax.numpy as jnp
from math import sin,cos
import matplotlib.pyplot as plt 
jax.config.update("jax_enable_x64", True)
```

```python
def f(x,y,z):
    return 2*(x**2) + 3*x*y - 4*z + jnp.sin(x) 

derivada = jax.grad(f, argnums=(0,1,2)) #transforma f em uma funcao que calcula gradientes
x = jnp.linspace(-2,2,100)   #derivada retorna uma tupla (df/dx, df/dy,df/dz), sendo assim precisamos calcular a derivada em cada ponto desejado separadamente
y = 1.0   #a funcao grad usa o backpropagation para calcular gradientes
z = 2.0
vet = jax.vmap(derivada,in_axes=(0, None, None), out_axes=0)
dfdx, dfdy, dfdz = vet(x,y,z)
plt.figure(figsize=(12, 8))
plt.plot(x, dfdx, label='Derivada parcial em x', color='blue')
plt.plot(x, dfdy, label='Derivada parcial em y', color='red', linestyle='--')
plt.plot(x, dfdz, label='Derivada parcial em z', color='green', linestyle=':')
plt.title('Derivadas Parciais')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
