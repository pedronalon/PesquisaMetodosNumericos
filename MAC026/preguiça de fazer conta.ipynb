```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(y,t):
    return -50*y + (50*(t**2)) + 2*t

h = 0.1
t = np.arange(0.0,1+h,h)

y = np.zeros(len(t))
y[0] = 1.0

for i in range(len(t) -1): 
    y[i+1] = y[i] + h*f(y[i],t[i])


dados  = {'t':t,'y':y}
df = pd.DataFrame(dados)
df
```
